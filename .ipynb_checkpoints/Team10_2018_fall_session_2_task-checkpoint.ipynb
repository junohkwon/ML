### Figure 5.2 - Validation Set Approach
Using Polynomial feature generation in scikit-learn<BR>
http://scikit-learn.org/dev/modules/preprocessing.html#generating-polynomial-features

```python
boston_df = pd.read_csv('./ISLR/Data/Boston.csv')
boston_x = boston_df.drop('medv', axis=1)
boston_y = boston_df.medv
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, BaggingRegressor, RandomForestRegressor, GradientBoostingRegressor

t_prop = 0.5
p_order = np.arange(1,11)
r_state = np.arange(0,10)

X, Y = np.meshgrid(p_order, r_state, indexing='ij')
Z = np.zeros((p_order.size,r_state.size))

# Generate 10 random splits of the dataset
for (i,j),v in np.ndenumerate(Z):
    X_train, X_test, y_train, y_test = train_test_split(boston_x, boston_y.ravel(), test_size=t_prop, random_state=i)
    regr = RandomForestRegressor(max_features=j+1, random_state=i)
    regr.fit(X_train, y_train)
    pred = regr.predict(X_test)

    Z[i,j]= mean_squared_error(y_test, pred)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4))

# Left plot (first split)
ax1.plot(X.T[0],Z.T[0], '-o')
ax1.set_title('Random split of the data set')

# Right plot (all splits)
ax2.plot(X,Z)
ax2.set_title('10 random splits of the data set')

for ax in fig.axes:
    ax.set_ylabel('Mean Squared Error')
    ax.set_ylim(10,35)
    ax.set_xlabel('Max Features')
    ax.set_xlim(0.5,10.5)
    ax.set_xticks(range(2,11,2));
```

### Figure 5.4

```python
# k-fold CV
folds = 10
X = list(range(1,11,1))
Z = []

    
for i in range(1,folds+1,1):
    regr = RandomForestRegressor(max_features=i, random_state=i)
    kf_10 = KFold(n_splits=folds, random_state=i)
    score = cross_val_score(regr, X_train, y_train, cv=kf_10, scoring='neg_mean_squared_error').mean()
    score = score * -1
    Z.append(score)
```

```python
plt.plot(X,Z, '-o')
plt.title('10-fold CV')
plt.ylabel('Mean Squared Error')
plt.xlabel('Degree of Max Feature')
```
